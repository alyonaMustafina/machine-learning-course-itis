```
import pandas as pd
import matplotlib.pyplot as plt
```

```
file = pd.read_csv('test.csv', sep=',')
file
```

```
fig = plt.figure(figsize=(100,30)) 
file.Sex.value_counts()
```

```
# alpha value used for blending 
file.Sex.value_counts().plot(kind="bar", alpha=0.3) 
```

```
file.Sex.value_counts().plot(kind="barh", alpha=0.3) 
```

```
file.plot(kind='scatter',x='Sex',y='Age',color='red')
plt.show()
```

```
file.groupby(['Pclass','Sex']).size().unstack()
```

```
file.groupby(['Pclass','Sex']).size().unstack().plot(kind='bar',stacked=True)
```

```
file.groupby(['Pclass','Sex']).size().plot(kind='bar')
plt.show()
```
